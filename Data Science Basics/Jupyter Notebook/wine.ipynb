```python
# Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
wine_data = pd.read_csv(
    'wine.csv', index_col = 0, keep_default_na=False
    ) 
```

### Histogram - Points per Wine Variety (Filtered)

```python
# Filter the wines
wine_data = pd.DataFrame(wine_data)
# Filter data for wines
selected_wines = [
    'Cabernet Sauvignon', 
    'Pinot Noir', 
    'Chardonnay'
    ]
# Filter the DataFrame
filtered_data = wine_data[
    wine_data['variety'].isin(selected_wines)
    ]
```

```python
# Create histogram plot of points for filtered wines
sns.histplot(
    y='points', 
    x='variety', 
    data=filtered_data
    )
# Add titles and labels
plt.title('Points per Wine Variety (Filtered)')
plt.ylabel('Points')
plt.xlabel('Wine Variety')
plt.show()
plt.close()
```
