# Data Visualisation

**Additional information**  
[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load data
car_data = pd.read_csv(
    'Cars93.csv', index_col = 0, keep_default_na=False
    ) 
```

### 1. Boxplot - Comparing Revs per Mile (Filtered Manufacturers)

```python
car_data = pd.DataFrame(car_data)
# Filter data for car manufacturers
selected_manufacturer = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
# Filter the DataFrame
filtered_data = car_data[car_data
                         ['Manufacturer'].isin(selected_manufacturer
                                               )]
```

```python
# Create the box plot
filtered_data.boxplot(
    column='Rev.per.mile', by='Manufacturer', grid=False
    )
# Add title and labels
plt.title("Rev per Mile by Manufacturer (Filtered)")
plt.suptitle("") 
plt.xlabel("Manufacturer")
plt.ylabel("Rev per Mile")
plt.show()
```

**Boxplot analysis:**  
The manufacturer with the car with the highest revs per mile is Toyota, with an outlier above 3400 revs per mile.
This is followed by Suzuki with one car with revs per mile just under 3400.

### 2. Histogram - Comparing MPG (City vs. Highway)

```python
# Histogram of MPG
sns.histplot(car_data['MPG.city'], 
             bins=20, 
             alpha=0.6, 
             label='MPG.city', 
             color='blue', 
             edgecolor=None
             )
sns.histplot(car_data['MPG.highway'], 
             bins=20, 
             alpha=0.6, 
             label='MPG.highway', 
             color='orange', 
             edgecolor=None
             )
# Add labels, title, and legend
plt.xlabel('Miles Per Gallon')
plt.ylabel('Number of Cars')
plt.title('MPG (City vs. Highway)')
plt.legend()
plt.show()
```

**Histogram analysis:**  
It is generally more fuel efficient to drive in the city.

### 3. Lineplot - Relationship between Wheelbase and Turning Circle

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_data)
# Add labels and title
plt.title('Wheelbase vs. Turning Circle')
plt.ylabel("Turning Circle")
plt.show()
```

**Lineplot analysis:**  
The larger the wheelbase, the greater the turning circle. Therefore these cars are more difficult to maneuver.

### 4. Barplot - Comparing Horsepower by Car Type

```python
# Barplot for mean horsepower
ave_horsepower = car_data.groupby(
    'Type')['Horsepower'].mean().reset_index()
ave_horsepower = ave_horsepower.sort_values(
    by='Horsepower',ascending=False)
sns.barplot(
    x='Type', 
    y='Horsepower', 
    data=ave_horsepower, 
    order=ave_horsepower['Type']
    )
# Add title
plt.title('Average Horsepower by Car Type')
plt.show()
```

**Barplot analysis:**  
On average, larger cars have more horsepower.
