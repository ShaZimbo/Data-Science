```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import the dataset
i_data = pd.read_csv("insurance.csv")
```

```python
# Explore the data
i_data.info()
i_data.describe()
```

```python
# Select and reshape data
x = i_data.iloc[:,0].values
y = i_data.iloc[:,6].values
y=y.reshape(-1, 1)
x=x.reshape(-1, 1)
```

```python
# Create an empty linear regression model 
age_charge = LinearRegression()

# Create the model
age_charge.fit(x,y)

# Use model to make a prediction
y_pred = age_charge.predict(x)

# Plot model and data, add labels
plt.scatter(x,y,color = 'b')
plt.title('Charges per Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Plot model and data with best fit line, add labels
plt.scatter(x,y,color = 'b')
plt.plot(x,age_charge.predict(x),color = 'r')
plt.title('Charges per Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
